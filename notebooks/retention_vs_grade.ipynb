```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import statsmodels.api as sm

file_input = "../results/Student Course Count.xlsx"
```

```python
df = pd.read_excel(file_input)
```

```python
plt.figure()
plt.scatter(df["Count"], df["Final Grade"], alpha=0.6)

plt.xlabel("Occurance Count")
plt.ylabel("GPA")
plt.title("Student GPA vs Retention")
plt.show()
```

```python
plt.figure()
moreThanFive = df[df["Count"] >=5]
plt.scatter(moreThanFive["Count"], moreThanFive["Final Grade"], alpha=0.6)

plt.xlabel("Occurance Count")
plt.ylabel("GPA")
plt.title("Student GPA vs Retention")
plt.show()
```

```python
plt.figure()
#get average for each arrival number then plot that
uniqueArrival = df["Count"].unique()
avg = {}
for count in uniqueArrival:
    avg [count] = df[df["Count"] == count]["Final Grade"].mean()
    
plt.scatter(avg.keys(), avg.values(), alpha=0.6)

plt.xlabel("Occurance Count")
plt.ylabel("GPA")
plt.title("Student GPA vs Retention")
plt.show()
```

```python
df = pd.read_excel(file_input)
x = df["Count"]
y = df["Final Grade"]

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

print(model.summary())
```

```python
plt.figure()
#get average for each arrival number then plot that
className = "INTRO PROB & STATS"

course = df[df["Course Name"] == className]
uniqueArrival = course["Count"].unique()
avg = {}
for count in uniqueArrival:
    avg [count] = course[course["Count"] == count]["Final Grade"].mean()
    
plt.scatter(avg.keys(), avg.values(), alpha=0.6)
plt.xlabel("Occurance Count")
plt.ylabel("GPA")
plt.title(f"{className} Student GPA AVG vs Retention")
plt.show()

plt.scatter(course["Count"], course["Final Grade"], alpha=0.6)
plt.xlabel("Occurance Count")
plt.ylabel("GPA")
plt.title(f"{className} Student GPA vs Retention")
plt.show()
```

```python
x = course["Count"]
y = course["Final Grade"]

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

print(model.summary())
```
